## NBA Player Analyis

```python
import pandas as pd
import numpy as np
```

```python
player_data = pd.read_csv('nba_player_data.csv', index_col="Unnamed: 0")
```

```python
player_data.columns
```

```python
for i in player_data.columns:
    try:
      print(i, len(np.unique(player_data[i].values)))
    except:
      continue
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Collect unique value counts
column_names = []
unique_counts = []

for i in player_data.columns:
    try:
        unique_count = len(np.unique(player_data[i].values))
        column_names.append(i)
        unique_counts.append(unique_count)
    except:
        continue

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(column_names, unique_counts, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Unique Values')
plt.title('Unique Value Count per Column in Player Data')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for Joel Embiid
embiid_data = player_data[player_data['player_name'] == 'Joel Embiid']

# Drop rows with missing age or pts
embiid_data = embiid_data.dropna(subset=['age', 'pts'])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(embiid_data['age'], embiid_data['pts'], marker='o', linestyle='-', color='royalblue')
plt.title("Joel Embiid: Age vs Points")
plt.xlabel("Age")
plt.ylabel("Points")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
Counter(player_data['team_abbreviation'])
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Count entries per team and sort from greatest to least
team_counts = Counter(player_data['team_abbreviation'])
sorted_counts = sorted(team_counts.items(), key=lambda x: x[1], reverse=True)

# Extract sorted teams and counts
teams = [team for team, count in sorted_counts]
counts = [count for team, count in sorted_counts]

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(teams, counts, color='mediumpurple', edgecolor='black')
plt.title("Number of Entries per NBA Team (Greatest to Least)")
plt.xlabel("Team Abbreviation")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Netflix exploratory analysis

```python
netflix_ratings = {}
current_key = None

with open('netflix_inspect.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        if line.endswith(':'):
            current_key = line[:-1]  # remove the colon
            # print(f"Key: {current_key}")
            netflix_ratings[current_key] = []
        elif current_key is not None:
            # print(f"Value: {line}")
            netflix_ratings[current_key].append(line)
```

```python
movie_titles = pd.read_csv('movie_titles.csv', header=None, index_col=0)
```

```python
movies = list(movie_titles[2].values)

title_dict = dict(zip(movie_titles.index, movie_titles[2]))
```

```python
netflix_in_both = netflix_titles[netflix_titles['title'].isin(movies)]
np.unique(netflix_in_both['type'])
```

```python
netflix_in_both['rating']
```

```python
import matplotlib.pyplot as plt

# Count each rating category
rating_counts = netflix_in_both['rating'].value_counts().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(rating_counts.index, rating_counts.values, color='lightseagreen', edgecolor='black')
plt.title('Distribution of Ratings in Netflix Shared Dataset')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data
titles = []
avg_scores = []

for key, value in netflix_ratings.items():
    title = title_dict[int(key)]
    scores = [int(i.split(',')[1]) for i in value]
    avg_score = np.mean(scores)

    titles.append(title)
    avg_scores.append(avg_score)

# Sort by average score (optional, for clearer viewing)
sorted_data = sorted(zip(titles, avg_scores), key=lambda x: x[1], reverse=True)
titles_sorted, scores_sorted = zip(*sorted_data)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(titles_sorted, scores_sorted, color='crimson', edgecolor='black')
plt.title("Average Netflix Rating per Title")
plt.xlabel("Movie Title")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
